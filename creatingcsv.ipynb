```python
import streamlit as st
from PIL import Image
import os
import json
from streamlit_option_menu import option_menu
import pandas as pd
import sqlite3
import plotly.express as px
```

```python
path='C:/Users/Admin/Desktop/pulse/data/aggregated/transaction/country/india/state/'
Agg_state_list=os.listdir(path)
clm={"State":[],"Year":[],"Quater":[],"transaction_type":[],"Transaction_count":[],"transaction_amount":[]}
```

```python
for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    #print(Agg_yr)
    for j in Agg_yr:
        p_j=p_i+j+'/'
        #print(p_i)
        Agg_list_yr=os.listdir(p_j)
        #print(Agg_list_yr)
        for k in Agg_list_yr:
            p_k=p_j+k
            #print(p_k)
            data=open(p_k,'r')
            #print(data)
            d=json.load(data)
            #print(d)
            for z in d['data']['transactionData']:
                name=z['name']
                count=z['paymentInstruments'][0]['count']
                amount=z['paymentInstruments'][0]['amount']
                clm["transaction_type"].append(name)
                clm["Transaction_count"].append(count)
                clm["transaction_amount"].append(amount)
                clm["State"].append(i)
                clm["Year"].append(j)
                clm["Quater"].append(int(k.strip('.json')))
df=pd.DataFrame(clm)
```

```python
df.to_csv("data aggregated transaction.csv")
```

```python
path2="C:/Users/Admin/Desktop/pulse/data/aggregated/user/country/india/state/"
Agg_state_list=os.listdir(path2)
clm1={"State":[],"Year":[],"Quater":[],"Brand":[],"Count_":[],"Percentage_":[]}
```

```python
for i in Agg_state_list:
    p_i=path2+i+'/'
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_list_year=os.listdir(p_j)
        for k in Agg_list_year:
            p_k=p_j+k
            data1=open(p_k,"r")
            b=json.load(data1)
            try:
                for w in b["data"]["usersByDevice"]:
                    brand_name=w["brand"]
                    count_=w["count"]
                    Allpercentage=w["percentage"]
                    clm1["Brand"].append(brand_name)
                    clm1["Count_"].append(count_)
                    clm1["Percentage_"].append(Allpercentage)
                    clm1["State"].append(i)
                    clm1["Year"].append(j)
                    clm1["Quater"].append(int(k.strip(".json")))
            except:
                pass
df1=pd.DataFrame(clm1)
df1.to_csv("data aggregater_user1.csv")  
#df1
```

```python
path3="C:/Users/Admin/Desktop/pulse/data/map/transaction/hover/country/india/state/"
map_state_list=os.listdir(path3)
col2={"State":[],"Year":[],"Quater":[],"district":[],"Count":[],"Amount":[]}
```

```python
for i in map_state_list:
    p_i=path3+i+'/'
    Map_yr=os.listdir(p_i)
    for j in Map_yr:
        p_j=p_i+j+'/'
        Map_list_year=os.listdir(p_j)
        for k in Map_list_year:
            p_k=p_j+k
            data=open(p_k,'r')
            d=json.load(data)
            try:
                for z in d["data"]["hoverDataList"]:
                    name_=z["name"]
                    count_=z["metric"][0]["count"]
                    Amount=z["metric"][0]["amount"]
                    col2["district"].append(name_)
                    col2["Count"].append(count_)
                    col2["Amount"].append(Amount)
                    col2["State"].append(i)
                    col2["Year"].append(j)
                    col2["Quater"].append(int(k.strip(".json")))
            except:
                pass
df2=pd.DataFrame(col2)
df2.to_csv("data map_transaction.csv")
```

```python
path3="C:/Users/Admin/Desktop/pulse/data/map/user/hover/country/india/state/"
Map_state_list=os.listdir(path3)
col3={"State":[],"Year":[],"Quater":[],"District":[],"RegisteredUsers":[]}
```

```python
for i in Map_state_list:
    p_i=path3+i+'/'
    Map_yr=os.listdir(p_i)
    for j in Map_yr:
        p_j=p_i+j+'/'
        Map_list_year=os.listdir(p_j)
        for k in Map_list_year:
            p_k=p_j+k
            data=open(p_k,"r")
            d=json.load(data)
            
            for u in d["data"]["hoverData"].items():
                district = u[0]
                registereduser = u[1]["registeredUsers"]
                col3["District"].append(district)
                col3["RegisteredUsers"].append(registereduser)
                col3['State'].append(i)
                col3['Year'].append(j)
                col3['Quater'].append(int(k.strip('.json')))
df_map_user = pd.DataFrame(col3)
df_map_user.to_csv("df_map_user.csv")
                
            
```
